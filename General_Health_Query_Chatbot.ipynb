# Task 4: General Health Query Chatbot

## Objective

Develop a chatbot that answers general health-related questions using a Hugging Face Large Language Model (FLAN-T5). The chatbot uses prompt engineering and safety filters to provide helpful responses while avoiding harmful medical advice.

```python
import transformers
print("Transformers Version:", transformers.__version__)
```

```python
from transformers import pipeline
```

```python
chatbot = pipeline(
    "text2text-generation",
    model="google/flan-t5-base"
)

print("✅ Model Loaded Successfully")
```

```python
unsafe_keywords = [
    "suicide",
    "kill myself",
    "self harm",
    "harm myself",
    "overdose"
]
```

```python
def safety_check(question):

    question = question.lower()

    for word in unsafe_keywords:
        if word in question:
            return False

    return True
```

```python
def health_chatbot(question):

    if not safety_check(question):
        return (
            "I cannot provide assistance for that request.\n"
            "Please contact a healthcare professional or emergency services."
        )

    prompt = (
        f"Answer this health question briefly and simply.\n"
        f"Do not diagnose diseases.\n"
        f"Do not prescribe medicine.\n\n"
        f"Question: {question}"
    )

    result = chatbot(
        prompt,
        max_new_tokens=80,
        do_sample=False
    )

    answer = result[0]["generated_text"].strip()

    answer += (
        "\n\nNote: This is general health information only. "
        "Please consult a healthcare professional for personal medical advice."
    )

    return answer
```

```python
print(health_chatbot("What causes a sore throat?"))
```

```python
print(health_chatbot("Is paracetamol safe for children?"))
```

```python
print(health_chatbot("How can I prevent dehydration?"))
```

```python
print(health_chatbot("What are the symptoms of the common cold?"))
```

```python
print(health_chatbot("I want to kill myself"))
```

```python
print("Health Chatbot")
print("Type 'exit' to quit.\n")

while True:

    question = input("You: ")

    if question.lower() == "exit":
        print("Bot: Goodbye!")
        break

    print("\nBot:")
    print(health_chatbot(question))
    print()
```

```python
print(health_chatbot("What foods are rich in Vitamin C?"))
```
